```python
import re
import matplotlib.pyplot as plt

# File path to the text file
file_path = './fps_params_tiles_results_shuffled.txt'

# Initialize lists to hold the extracted data
frames = []
fps = []
avg_tiles_hit = []

# Regular expression to match the data pattern
pattern = r'Frame_(\d+): FPS: ([\d\.]+),.*Avg Tiles Hit: ([\d\.]+)'

# Read the file and extract the data
with open(file_path, 'r') as file:
    for line in file:
        match = re.search(pattern, line)
        if match:
            frames.append(int(match.group(1)))
            fps.append(float(match.group(2)))
            avg_tiles_hit.append(float(match.group(3)))

# Create subplots for FPS and Avg Tiles Hit
plt.figure(figsize=(12, 6))

# Plot FPS
plt.subplot(1, 2, 1)
plt.plot(frames, fps, marker='o')
plt.title('FPS over Frames')
plt.xlabel('Frame')
plt.ylabel('FPS')

# Plot Avg Tiles Hit
plt.subplot(1, 2, 2)
plt.plot(frames, avg_tiles_hit, marker='o', color='orange')
plt.title('Avg Tiles Hit over Frames')
plt.xlabel('Frame')
plt.ylabel('Avg Tiles Hit')

# Save the plots as an image file
output_image_path = './fps_avg_tiles_plot.png'
plt.tight_layout()
plt.savefig(output_image_path)

# Show the plots
plt.show()

print(f"Plot saved as {output_image_path}")
```
