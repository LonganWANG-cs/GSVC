```python
import re
import matplotlib.pyplot as plt

# File path to the text file
file_path = './fps_params_tiles_results_shuffled.txt'

# Initialize lists to hold the extracted data
frames = []
fps = []
avg_tiles_hit = []

# Regular expression to match the data pattern
pattern = r'Frame_(\d+): FPS: ([\d\.]+),.*Avg Tiles Hit: ([\d\.]+)'

# Read the file and extract the data
with open(file_path, 'r') as file:
    for line in file:
        match = re.search(pattern, line)
        if match:
            frames.append(int(match.group(1)))
            fps.append(float(match.group(2)))
            avg_tiles_hit.append(float(match.group(3)))

# Plot FPS and Avg Tiles Hit similar to plot_gaussian_vs_metrics style
plt.figure(figsize=(15, 10))

# FPS scatter and line plot
plt.subplot(2, 1, 1)
plt.scatter(frames, fps, color='blue', alpha=0.7, s=50, label='FPS', edgecolor='black')
plt.plot(frames, fps, 'b-', alpha=0.5)

# Add labels and title
plt.xlabel('Frame', fontsize=14, fontweight='bold')
plt.ylabel('FPS', fontsize=14, fontweight='bold')
plt.title('Frame vs FPS', fontsize=16, fontweight='bold')

# Add grid, xticks, and legend
plt.xticks(range(min(frames), max(frames)+1, int(len(frames) / 10)), fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
#plt.legend()

# Avg Tiles Hit scatter and line plot
plt.subplot(2, 1, 2)
plt.scatter(frames, avg_tiles_hit, color='orange', alpha=0.7, s=50, label='Avg Tiles Hit', edgecolor='black')
plt.plot(frames, avg_tiles_hit, 'orange', alpha=0.5)

# Add labels and title
plt.xlabel('Frame', fontsize=14, fontweight='bold')
plt.ylabel('Avg Tiles Hit', fontsize=14, fontweight='bold')
plt.title('Frame vs Avg Tiles Hit', fontsize=16, fontweight='bold')

# Add grid, xticks, and legend
plt.xticks(range(min(frames), max(frames)+1, int(len(frames) / 10)), fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
#plt.legend()

# Adjust layout and save the figure
plt.tight_layout(pad=3.0)
output_image_path = './fps_avg_tiles_plot.png'
plt.savefig(output_image_path, dpi=300)

# Show the plot
plt.show()

print(f"Plot saved as {output_image_path}")
```
